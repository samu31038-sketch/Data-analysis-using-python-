```python

import pandas as pd

data = {
    "Region": ["South", "North", "East", "West", "South", "North"],
    "Products": ["Labtop", "Mobile", "Tablet", "Laptop", "Mobile", "Tablet"],
    "Sales": [7000, 8000, 9000, 4000, 60000, 31000]
}

df = pd.DataFrame(data)
print(df)

south_sales = df[df["Region"] == "South"]

print(south_sales)
```
